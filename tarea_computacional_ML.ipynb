## XGBoost & SVM

Se usó una partición estratificada 80-20 para todos los modelos, asegurando una evaluación consistente y una comparación justa de los resultados.

```python
from sklearn.model_selection import GridSearchCV, train_test_split, KFold
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, mean_absolute_percentage_error
from xgboost import XGBRegressor
from sklearn.svm import SVR
import numpy as np

# Dividir el dataframe en X (características) e Y (variable objetivo)
X = df.drop(columns=['Calories_Burned'])
Y = df['Calories_Burned']

# Dividir en conjuntos de entrenamiento y prueba La misma para todos los modelos
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Normalizar los datos (solo X, no Y) **IMPORTANTE**
mean, std = X_train.mean(axis=0), X_train.std(axis=0)
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std# Normalizar los datos (solo X, no Y) **IMPORTANTE**
mean, std = X_train.mean(axis=0), X_train.std(axis=0)
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std

# Definir validación cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Función para calcular métricas
def evaluate_model_with_cv(model):
    mae_list, r2_list, mape_list, rmse_list = [], [], [], []
    for train_index, test_index in kf.split(X_train):
        X_train_fold, X_val_fold = X_train.iloc[train_index], X_train.iloc[test_index]
        Y_train_fold, Y_val_fold = Y_train.iloc[train_index], Y_train.iloc[test_index]

        # Entrenar el modelo
        model.fit(X_train_fold, Y_train_fold)

        # Predicciones
        Y_pred = model.predict(X_val_fold)

        # Calcular métricas
        mae_list.append(mean_absolute_error(Y_val_fold, Y_pred))
        r2_list.append(r2_score(Y_val_fold, Y_pred))
        mape_list.append(mean_absolute_percentage_error(Y_val_fold, Y_pred))
        rmse_list.append(mean_squared_error(Y_val_fold, Y_pred, squared=False))

    # Promedio de métricas
    return {
        "Mean MAE": np.mean(mae_list),
        "Mean R²": np.mean(r2_list),
        "Mean MAPE": np.mean(mape_list) * 100,
        "Mean RMSE": np.mean(rmse_list),
    }

# Optimización de hiperparámetros con GridSearchCV

## 1. Optimización de XGBoost
xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

xgb_grid_search = GridSearchCV(
    XGBRegressor(random_state=42),
    param_grid=xgb_param_grid,
    cv=3,
    scoring='neg_mean_absolute_error',
    verbose=1
)

xgb_grid_search.fit(X_train, Y_train)

# Mejor modelo de XGBoost
best_xgb_model = xgb_grid_search.best_estimator_
print("Best parameters for XGBoost:", xgb_grid_search.best_params_)

## 2. Optimización de SVM
svm_param_grid = {
    'C': [1, 10, 100],
    'gamma': ['scale', 'auto', 0.1, 0.01],
    'kernel': ['rbf', 'poly']
}

svm_grid_search = GridSearchCV(
    SVR(),
    param_grid=svm_param_grid,
    cv=3,
    scoring='neg_mean_absolute_error',
    verbose=1
)

svm_grid_search.fit(X_train, Y_train)

# Mejor modelo de SVM
best_svm_model = svm_grid_search.best_estimator_
print("Best parameters for SVM:", svm_grid_search.best_params_)

# Evaluar ambos modelos optimizados
xgb_metrics = evaluate_model_with_cv(best_xgb_model)
svm_metrics = evaluate_model_with_cv(best_svm_model)

# Mostrar resultados
print("\nOptimized XGBoost Metrics:")
for metric, value in xgb_metrics.items():
    print(f"{metric}: {value:.2f}")

print("\nOptimized SVM Metrics:")
for metric, value in svm_metrics.items():
    print(f"{metric}: {value:.2f}")
```


Para optimizar la red neuronal, comenzamos con una arquitectura básica y la fuimos incrementando en complejidad de manera progresiva, ajustando gradualmente sus parámetros y capas. Su artitectura e hiperparametros se explicaran a continuación


```python
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Nadam
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras import models, layers
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# EarlyStopping con mayor paciencia
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# ReduceLROnPlateau para ajustar el learning rate
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-7
)


# Construcción del modelo mejorado
def build_improved_model():
    model = models.Sequential()

    # Primera capa oculta
    model.add(layers.Dense(256, input_shape=(X_train.shape[1],)))
    model.add(layers.LeakyReLU(alpha=0.1))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))

    # Segunda capa oculta
    model.add(layers.Dense(128, kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))
    model.add(layers.LeakyReLU(alpha=0.1))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))

    # Tercera capa oculta
    model.add(layers.Dense(64))
    model.add(layers.LeakyReLU(alpha=0.1))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))

    # Cuarta capa oculta
    model.add(layers.Dense(32))
    model.add(layers.LeakyReLU(alpha=0.1))
    model.add(layers.BatchNormalization())

    # Capa de salida
    model.add(layers.Dense(1))

    # Compilación del modelo
    model.compile(
        optimizer=Nadam(learning_rate=0.001),
        loss='huber',
        metrics=['mae']
    )
    return model


# Crear y entrenar el modelo
model = build_improved_model()

history = model.fit(
    X_train, Y_train,
    epochs=200,
    batch_size=16,
    validation_data=(X_test, Y_test),
    callbacks=[early_stopping, reduce_lr]
)

# Visualizar las curvas de entrenamiento
plt.figure(figsize=(12, 6))
plt.plot(history.history['mae'], label='MAE de entrenamiento')
plt.plot(history.history['val_mae'], label='MAE de validación')
plt.xlabel('Épocas')
plt.ylabel('MAE')
plt.legend()
plt.title('MAE durante el entrenamiento y la validación')
plt.show()
```

La red neuronal tiene cuatro capas ocultas con unidades decrecientes (256, 128, 64 y 32), activación LeakyReLU (alpha=0.1), Batch Normalization en cada capa y Dropout (0.3) en las tres primeras para prevenir el sobreajuste. La segunda capa incluye regularización L1/L2. Se compila con el optimizador Nadam, que combina las ventajas de RMSprop y momentum para un aprendizaje más eficiente, y la pérdida Huber, que es robusta frente a outliers. Además, se utilizan EarlyStopping para detener el entrenamiento cuando el modelo deje de mejorar y ReduceLROnPlateau para ajustar dinámicamente la tasa de aprendizaje, mejorando la estabilidad del entrenamiento.

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_squared_error, mean_absolute_error

# Hacer predicciones
Y_pred = model.predict(X_test)

# Calcular métricas
mape = mean_absolute_percentage_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
rmse = mean_squared_error(Y_test, Y_pred, squared=False)
mae = mean_absolute_error(Y_test, Y_pred)

# Imprimir métricas
print(f"MAPE: {mape * 100:.2f}%")
print(f"R^2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

k = 4  # Número de particiones
num_val_samples = len(X_train) // k  # Tamaño de cada partición
num_epochs = 200
all_mae_scores = []  # Guardar los MAE de cada fold

for i in range(k):
    print(f'Processing fold #{i + 1}')

    # Dividir datos para validación y entrenamiento
    val_data = X_train[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = Y_train[i * num_val_samples: (i + 1) * num_val_samples]

    partial_train_data = np.concatenate(
        [X_train[:i * num_val_samples],
         X_train[(i + 1) * num_val_samples:]],
        axis=0)
    partial_train_targets = np.concatenate(
        [Y_train[:i * num_val_samples],
         Y_train[(i + 1) * num_val_samples:]],
        axis=0)

    # Construir el modelo
    model = build_improved_model()

    # Entrenar el modelo
    history = model.fit(
        partial_train_data, partial_train_targets,
        epochs=num_epochs,
        batch_size=16,
        validation_data=(val_data, val_targets),
        callbacks=[reduce_lr],
        verbose=0  # Para reducir el texto durante el entrenamiento
    )

    # Evaluar el modelo en los datos de validación
    val_predictions = model.predict(val_data)
    val_mae = mean_absolute_error(val_targets, val_predictions)
    all_mae_scores.append(val_mae)

# Resultados de validación cruzada
print(f"Validation MAE scores for each fold: {all_mae_scores}")
print(f"Mean MAE across folds: {np.mean(all_mae_scores):.4f}")

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Melt del DataFrame para facilitar el plotting
metrics_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Value')

# Excluir R² de la gráfica para que las escalas sean coherentes
metrics_melted_no_r2 = metrics_melted[metrics_melted['Metric'] != 'Mean R²']

# Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Metric', y='Value', hue='Model', data=metrics_melted_no_r2)
plt.title('Comparación de Métricas entre Modelos')
plt.ylabel('Valor de la Métrica')
plt.legend(loc='upper right')
plt.show()
```

```python
# Filtrar solo el R²
metrics_r2 = metrics_melted[metrics_melted['Metric'] == 'Mean R²']

# Gráfico de barras para R²
plt.figure(figsize=(6, 4))
sns.barplot(x='Model', y='Value', data=metrics_r2)
plt.title('Comparación del R² entre Modelos')
plt.ylabel('Valor de R²')
plt.ylim(0.95, 1.00)  # Ajustar límites para resaltar diferencias
plt.show()
```

Basándonos en las métricas de desempeño (MAE, R², MAPE, y RMSE), el mejor modelo es XGBoost.

## ¿Seguiría siendo el modelo XGBoost el mejor modelo si consideramos la interpretabilidad?

XGBoost sigue siendo el mejor modelo al considerar la interpretabilidad, gracias a herramientas como SHAP y Feature Importance que permiten entender las contribuciones de cada variable.
Además, combina explicabilidad con el mejor desempeño en todas las métricas (MAE, R², MAPE, RMSE). SVM tiene una capacidad explicativa limitada en regresión, especialmente con kernels no lineales. Por lo tanto, XGBoost ofrece el mejor equilibrio entre precisión y explicabilidad.


## Conclusiones

Aunque XGBoost es la mejor opción actual, las redes neuronales tienen una ventaja significativa si se aumenta el volumen de datos, se introducen nuevas variables o se complejiza el problema.
